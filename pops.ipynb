## POP's cumulated by preys:

```python
%%capture capt
url = 'http://nextatlas.com/corporate-test/service/activities/now/52879cd31ea3e8ba20f34689'
acts_preys = get_NA_data(url)
```

```python
from collections import Counter
cum_pops = Counter()
#for prey_data in acts_preys['examples_activities']:
for prey_data in acts_preys:
    prey_votes= prey_data['votes']
    pops = [vote['pops'] for vote in prey_votes]
    #tss = [datetime.utcfromtimestamp( vote['timestamp']/1000 ) for vote in prey_votes]
    tss = [datetime.utcfromtimestamp( vote['timestamp'] ) for vote in prey_votes]
    cum_pops.update(dict(zip(tss, pops)))

dates_cum_preys, pops_cum_preys = zip(*sorted(cum_pops.iteritems()))
pops_cum_preys = np.cumsum(pops_cum_preys)
pylab.figure(figsize=(18,4))
pylab.plot(dates_cum_preys, pops_cum_preys,'.-')
pylab.show()
```

## comparation of trends and preys

```python
pylab.figure(figsize=(18,4))
pylab.title('trend activities')
pylab.plot(dates_trend, pops_trend,'.-')
pylab.plot(dates_cum_preys, pops_cum_preys,'.-', alpha=0.5)
pylab.ylabel("PoP's")
pylab.show()
```
